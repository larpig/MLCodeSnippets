```python
import numpy as np
import random
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

# Continuous Probability Distributions

```python
# ===========================================================================================
# Continuous Probability Distributions
#
# Uniform distribution - https://en.wikipedia.org/wiki/Uniform_distribution_(continuous)
# Triangular distribution - https://en.wikipedia.org/wiki/Triangular_distribution
# Normal distribution - https://en.wikipedia.org/wiki/Normal_distribution
# Exponential distribution - https://en.wikipedia.org/wiki/Exponential_distribution
# Gamma distribution - https://en.wikipedia.org/wiki/Gamma_distribution
# Beta distribution - https://en.wikipedia.org/wiki/Beta_distribution
# ===========================================================================================

nup = 10000

low = 0
high = 10
mode = 7
mu = 5
sigma = 2
lambd = 1/mu
alpha = 6
beta  = 3

# https://docs.python.org/3/library/random.html
x0 = [random.uniform(low, high) for i in range(nup)]
x1 = [random.triangular(low, high, mode) for i in range(nup)]
x2 = [random.gauss(mu, sigma) for i in range(nup)] # This is slightly faster than the normalvariate() function
x3 = [random.expovariate(lambd) for i in range(nup)]
x4 = [random.gammavariate(alpha, beta) for i in range(nup)]
x5 = [random.betavariate(alpha, beta) for i in range(nup)]

fig, axes = plt.subplots(nrows=2,ncols=3, figsize=(10, 6))
plt.subplots_adjust(left=.1, right=.98, bottom=.001, top=.85, wspace=0.3, hspace=.4)
fig.suptitle('Continuous Probability Distributions')
sns.distplot(x0,hist=True, hist_kws={"histtype": "step", "linewidth": 4, "alpha": 0.8, "color": "b"},                  
                kde=True, kde_kws={"color": "k", "lw": 2, "label": "KDE"}, ax=axes[0,0])
axes[0,0].set_xlabel('Uniform Distribution \n low=%.1f, high=%.0f' % (low,high),fontdict={"fontsize":12})
sns.distplot(x1,hist=True, hist_kws={"histtype": "step", "linewidth": 4, "alpha": 0.8, "color": "b"},                  
                kde=True, kde_kws={"color": "k", "lw": 2, "label": "KDE"}, ax=axes[0,1])
axes[0,1].set_xlabel('Triangular distribution \n low=%.1f, high=%.0f, mode=%.0f' % (low,high, mode),fontdict={"fontsize":12})
sns.distplot(x2,hist=True, hist_kws={"histtype": "step", "linewidth": 4, "alpha": 0.8, "color": "b"},                  
                kde=True, kde_kws={"color": "k", "lw": 2, "label": "KDE"}, ax=axes[0,2])
axes[0,2].set_xlabel('Normal distribution \n mu=%.0f, sigma=%.0f' % (mu,sigma),fontdict={"fontsize":12})
sns.distplot(x3,hist=True, hist_kws={"histtype": "step", "linewidth": 4, "alpha": 0.8, "color": "b"},                  
                kde=True, kde_kws={"color": "k", "lw": 2, "label": "KDE"}, ax=axes[1,0])
axes[1,0].set_xlabel('Exponential Distribution \n lambda=%.2f (1/mu, mu=%.0f)' % (lambd,mu),fontdict={"fontsize":12})
sns.distplot(x4,hist=True, hist_kws={"histtype": "step", "linewidth": 4, "alpha": 0.8, "color": "b"},                  
                kde=True, kde_kws={"color": "k", "lw": 2, "label": "KDE"}, ax=axes[1,1])
axes[1,1].set_xlabel('Gamma distribution \n alpha=%.1f, beta=%.0f' % (alpha,beta),fontdict={"fontsize":12})
sns.distplot(x5,hist=True, hist_kws={"histtype": "step", "linewidth": 4, "alpha": 0.8, "color": "b"},                  
                kde=True, kde_kws={"color": "k", "lw": 2, "label": "KDE"}, ax=axes[1,2])
axes[1,2].set_xlabel('Beta distribution \n alpha=%.1f, beta=%.0f' % (alpha,beta),fontdict={"fontsize":12})
plt.show()
```

# Discrete Probability Distributions

```python
# ===========================================================================================
# Discrete Probability Distributions
#
# Uniform distribution - https://en.wikipedia.org/wiki/Discrete_uniform_distribution
# Bernoulli distribution - https://en.wikipedia.org/wiki/Bernoulli_distribution
# Binomial distribution - https://en.wikipedia.org/wiki/Binomial_distribution
# Poisson distribution - https://en.wikipedia.org/wiki/Poisson_distribution
# Exponential distribution - https://en.wikipedia.org/wiki/Exponential_distribution
# Hypergeometric distribution - https://en.wikipedia.org/wiki/Hypergeometric_distribution
# Beta-binomial - https://en.wikipedia.org/wiki/Beta-binomial_distribution
# Negative Binomial Distribution - https://en.wikipedia.org/wiki/Negative_binomial_distribution
# ===========================================================================================

nup = 10000
low = 0
high = 10
p=0.6
n=10
mu=3
lambd = 1/mu

# https://docs.scipy.org/doc/scipy/reference/tutorial/stats/discrete.html
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.rv_discrete.html
x0 = stats.randint.rvs(size=nup, low=low, high=high)
x0_unique, x0_counts = np.unique(x0, return_counts=True)
x0_dist = stats.rv_discrete(name='uniform', values=(x0_unique, x0_counts/nup))

x1 = stats.bernoulli.rvs(size=nup,p=p)
x1_unique, x1_counts = np.unique(x1, return_counts=True)
x1_dist = stats.rv_discrete(name='bernoulli', values=(x1_unique, x1_counts/nup))

x2 = stats.binom.rvs(n=n,p=p,size=nup)
x2_unique, x2_counts = np.unique(x2, return_counts=True)
x2_dist = stats.rv_discrete(name='binomial', values=(x2_unique, x2_counts/nup))

x3 = stats.poisson.rvs(mu=mu, size=nup)
x3_unique, x3_counts = np.unique(x3, return_counts=True)
x3_dist = stats.rv_discrete(name='poisson', values=(x3_unique, x3_counts/nup))

x4 = stats.geom.rvs(size=nup,p=p)
x4_unique, x4_counts = np.unique(x4, return_counts=True)
x4_dist = stats.rv_discrete(name='geometric', values=(x4_unique, x4_counts/nup))

x5 = stats.planck.rvs(size=nup,lambda_=lambd) # discrete exponential
x5_unique, x5_counts = np.unique(x5, return_counts=True)
x5_dist = stats.rv_discrete(name='geometric', values=(x5_unique, x5_counts/nup))

# Suppose we have a collection of 20 animals, of which 7 are dogs. 
# Then if we want to know the probability of finding a given number of dogs 
# if we choose at random 12 of the 20 animals.
# The output is the number of dogs in our group of chosen animals.
[M_elemen, n_desire, N_draw] = [20, 7, 12]
x6 = stats.hypergeom.rvs(size=nup,M=M_elemen, n=n_desire, N=N_draw)
x6_unique, x6_counts = np.unique(x6, return_counts=True)
x6_dist = stats.rv_discrete(name='hypergeometric', values=(x6_unique, x6_counts/nup))

# The beta-binomial distribution is the binomial distribution in which the 
# probability of success at each of n trials is not fixed but randomly drawn 
# from a beta distribution. It is frequently used in Bayesian statistics, 
# empirical Bayes methods and classical statistics to capture overdispersion 
# in binomial type distributed data
alpha = 6
beta  = 3
x7 = stats.betabinom.rvs(n=n, a=alpha, b=beta,size=nup)
x7_unique, x7_counts = np.unique(x7, return_counts=True)
x7_dist = stats.rv_discrete(name='beta-binomial', values=(x7_unique, x7_counts/nup))

# Negative binomial distribution that models the number k of failures in a sequence 
# of independent and identically distributed Bernoulli trials before a specified 
# number n of successes occurs, given the probability p of a single success.
nsuc = 7
x8 = stats.nbinom.rvs(n=nsuc,p=p,size=nup)
x8_unique, x8_counts = np.unique(x8, return_counts=True)
x8_dist = stats.rv_discrete(name=' negative binomial', values=(x8_unique, x8_counts/nup))

fig, axes = plt.subplots(nrows=3,ncols=3, figsize=(10, 9))
plt.subplots_adjust(left=.1, right=.98, bottom=.001, top=.85, wspace=0.3, hspace=.4)
fig.suptitle('Discrete Probability Distributions')
axes[0,0].plot(x0_unique, x0_dist.pmf(x0_unique), 'ro', ms=4, mec='r')
axes[0,0].vlines(x0_unique, 0, x0_dist.pmf(x0_unique), colors='r', lw=2)
axes[0,0].set_xlabel('Uniform Distribution \n low=%.1f, high=%.0f' % (low,high),fontdict={"fontsize":12})
axes[0,1].plot(x1_unique, x1_dist.pmf(x1_unique), 'ro', ms=4, mec='r')
axes[0,1].vlines(x1_unique, 0, x1_dist.pmf(x1_unique), colors='r', lw=2)
axes[0,1].set_xlabel('Bernoulli Distribution \n p=%.1f' % p,fontdict={"fontsize":12})
axes[0,2].plot(x2_unique, x2_dist.pmf(x2_unique), 'ro', ms=4, mec='r')
axes[0,2].vlines(x2_unique, 0, x2_dist.pmf(x2_unique), colors='r', lw=2)
axes[0,2].set_xlabel('Binomial Distribution \n p=%.1f, n=%.0f' % (p,n),fontdict={"fontsize":12})
axes[1,0].plot(x3_unique, x3_dist.pmf(x3_unique), 'ro', ms=4, mec='r')
axes[1,0].vlines(x3_unique, 0, x3_dist.pmf(x3_unique), colors='r', lw=2)
axes[1,0].set_xlabel('Poisson Distribution \n mu=%.1f' % mu,fontdict={"fontsize":12})
axes[1,1].plot(x4_unique, x4_dist.pmf(x4_unique), 'ro', ms=4, mec='r')
axes[1,1].vlines(x4_unique, 0, x4_dist.pmf(x4_unique), colors='r', lw=2)
axes[1,1].set_xlabel('Geometric Distribution \n p=%.1f' % p,fontdict={"fontsize":12})
axes[1,2].plot(x5_unique, x5_dist.pmf(x5_unique), 'ro', ms=4, mec='r')
axes[1,2].vlines(x5_unique, 0, x5_dist.pmf(x5_unique), colors='r', lw=2)
axes[1,2].set_xlabel('Exponential Distribution \n lambda=%.2f (1/mu, mu=%.0f)' % (lambd,mu),fontdict={"fontsize":12})
axes[2,0].plot(x6_unique, x6_dist.pmf(x6_unique), 'ro', ms=4, mec='r')
axes[2,0].vlines(x6_unique, 0, x6_dist.pmf(x6_unique), colors='r', lw=2)
axes[2,0].set_xlabel('Hypergeometric Distribution \n M=%.0f, n=%.0f, N=%.0f' % (M_elemen,n_desire,N_draw),fontdict={"fontsize":12})
axes[2,1].plot(x7_unique, x7_dist.pmf(x7_unique), 'ro', ms=4, mec='r')
axes[2,1].vlines(x7_unique, 0, x7_dist.pmf(x7_unique), colors='r', lw=2)
axes[2,1].set_xlabel('Beta-binomial Distribution \n n=%.0f, alpha=%.0f, beta=%.0f' % (n,alpha,beta),fontdict={"fontsize":12})
axes[2,2].plot(x8_unique, x8_dist.pmf(x8_unique), 'ro', ms=4, mec='r')
axes[2,2].vlines(x8_unique, 0, x8_dist.pmf(x8_unique), colors='r', lw=2)
axes[2,2].set_xlabel('Negative Binomial Distribution \n p=%.1f, n=%.0f' % (p,nsuc),fontdict={"fontsize":12})
plt.show()
```

# Simulation of an experiment to get the opt-in of clients

Experiment consists of sending a communitation for a specific number of customers with a link of a landing page (LP), being that this LP contains a request of opt-in for those customers that opened it.

It's assumed that:
- The number of participants of the experiment is equal to n;
- The minimum desired number of obtained opt-in is equal to n_suc;
- The communication channel presents a click true rate (CTR) equals to p1;
- The LP presents an opt-in rate (based on only the clients that opened it) equals to p2.


```python
p1 = 1/100
p2 = 20/100
p = p1*p2
n = 15000
print('Hypothesis:')
print('Click True Rate (CTR) of Landing Page (LP): %.4f' % p1)
print('Opt-in rate for who opened the Landing Page (LP): %.4f' % p2)
print('Opt-in rate for who participated of the experiment: %.4f' % p)

print('\nResults:')
print('Number of opt-in for %.d participants based on opt-in rate: %.d' % (n, n*p))

x1 = stats.binom.rvs(n=n,p=p,size=nup)
x1_min = min(x1)
x1_avg = np.mean(x1)
x1_max = max(x1)
print('Number of opt-in for %.d participants considering a binominal distribution: (min=%.d, avg=%.d, max=%.d)' % (n, x1_min, x1_avg, x1_max))

alpha = int(nup*p)
beta = int(nup*(1-p))
x2 = stats.betabinom.rvs(n=n, a=alpha, b=beta,size=nup)
x2_min = min(x2)
x2_avg = np.mean(x2)
x2_max = max(x2)
print('Number of opt-in for %.d participants considering a beta-binominal distribution: (min=%.d, avg=%.d, max=%.d)' % (n, x2_min, x2_avg, x2_max))
```

```python
n_suc = 30
n_sim = 1000
x3 = stats.nbinom.rvs(n=n_suc,p=p,size=n_sim)

x3_min = min(x3)
x3_avg = np.mean(x3)
x3_max = max(x3)
x3_std = np.std(x3)

print('Number of rejections to achieve %.d opt-in based on an experiment with %.d draws: \n\t min=%.d, \n\t avg=%.0f, \n\t max=%.d \n' % (n_suc, n_sim, x3_min, x3_avg, x3_max))

print('Assuming a Normal Distribution with mean = %.0f and standard deviation = %.0f, one has:' % (x3_avg, x3_std))
print('\t Approximately 68%% of the number of rejections belong to the interval (%.d, %.d)' % (x3_avg-1*x3_std, x3_avg+1*x3_std))
print('\t Approximately 95%% of the number of rejections belong to the interval (%.d, %.d)' % (x3_avg-2*x3_std, x3_avg+2*x3_std))
print('\t Approximately 99%% of the number of rejections belong to the interval (%.d, %.d)' % (x3_avg-3*x3_std, x3_avg+3*x3_std))

fig, ax = plt.subplots(nrows=1,ncols=1, figsize=(3, 3))
fig.suptitle('Number of rejections to achieve %.d opt-in in %.d draws' % (n_suc,n_sim),fontdict={"fontsize":12})
ax.hist(x3)
plt.show()
```
